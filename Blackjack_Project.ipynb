## Python Blackjack

### Rules:
`1. ` The game will have two players: the Dealer and the Player. The game will start off with a deck of 52 cards. The 52 cards will consist of 4 different suits: Clubs, Diamonds, Hearts and Spades. For each suit, there will be cards numbered 1 through 13. <br>

**Note: No wildcards will be used in the program**

`2. ` When the game begins, the dealer will shuffle the deck of cards, making them randomized. After the dealer shuffles, it will deal the player 2 cards and will deal itself 2 cards from. The Player should be able to see both of their own cards, but should only be able to see one of the Dealer's cards.

`3. ` The objective of the game is for the Player to count their cards after they're dealt. If they're not satisfied with the number, they have the ability to 'Hit'. A hit allows the dealer to deal the Player one additional card. The Player can hit as many times as they'd like as long as they don't 'Bust'. A bust is when the Player is dealt cards that total more than 21.

`4. ` If the dealer deals the Player cards equal to 21 on the **first** deal, the Player wins. This is referred to as Blackjack. Blackjack is **NOT** the same as getting cards that equal up to 21 after the first deal. Blackjack can only be attained on the first deal.

`5. ` The Player will never see the Dealer's hand until the Player chooses to 'stand'. A Stand is when the player tells the dealer to not deal it anymore cards. Once the player chooses to Stand, the Player and the Dealer will compare their hands. Whoever has the higher number wins. Keep in mind that the Dealer can also bust.

```python
import random

# ---- Card Class ----
class Card:
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank  # 1 to 13

    def __str__(self):
        name = {1: "Ace", 11: "Jack", 12: "Queen", 13: "King"}.get(self.rank, str(self.rank))
        return f"{name} of {self.suit}"

    def value(self):
        if self.rank > 10:
            return 10
        elif self.rank == 1:
            return 11  # Ace starts as 11
        return self.rank

# ---- Deck Class ----
class Deck:
    def __init__(self):
        suits = ["Clubs", "Diamonds", "Hearts", "Spades"]
        self.cards = [Card(suit, rank) for suit in suits for rank in range(1, 14)]

    def shuffle(self):
        random.shuffle(self.cards)

    def deal(self):
        return self.cards.pop()

# ---- Hand Class ----
class Hand:
    def __init__(self):
        self.cards = []

    def add_card(self, card):
        self.cards.append(card)

    def get_value(self):
        total = sum(card.value() for card in self.cards)
        # Handle Aces (can be 1 or 11)
        aces = sum(1 for card in self.cards if card.rank == 1)
        while total > 21 and aces:
            total -= 10  # Ace as 1 instead of 11
            aces -= 1
        return total

    def __str__(self):
        return ', '.join(str(card) for card in self.cards)

# ---- Player Class ----
class Player:
    def __init__(self, name="Player"):
        self.name = name
        self.hand = Hand()

    def hit(self, deck):
        self.hand.add_card(deck.deal())

    def show_hand(self):
        print(f"{self.name}'s hand: {self.hand} (Value: {self.hand.get_value()})")

# ---- Dealer Class ----
class Dealer(Player):
    def __init__(self):
        super().__init__("Dealer")

    def show_initial_hand(self):
        print(f"Dealer shows: {self.hand.cards[0]} and [Hidden]")

# ---- Game Class ----
class BlackjackGame:
    def __init__(self):
        self.deck = Deck()
        self.deck.shuffle()
        self.player = Player()
        self.dealer = Dealer()

    def start(self):
        print("\n=== Starting Blackjack Game ===")
        # Initial deal
        for _ in range(2):
            self.player.hit(self.deck)
            self.dealer.hit(self.deck)

        self.player.show_hand()
        self.dealer.show_initial_hand()

        # Check for initial Blackjack
        if self.player.hand.get_value() == 21:
            print("Blackjack! Player wins!")
            return

        # Player's turn
        while True:
            choice = input("Do you want to (h)it or (s)tand? ").lower()
            if choice == "h":
                self.player.hit(self.deck)
                self.player.show_hand()
                if self.player.hand.get_value() > 21:
                    print("Bust! Dealer wins.")
                    return
            elif choice == "s":
                break
            else:
                print("Invalid input. Please enter 'h' or 's'.")

        # Dealer's turn
        print("\n=== Dealer's Turn ===")
        self.dealer.show_hand()
        while self.dealer.hand.get_value() < 17:
            print("Dealer hits...")
            self.dealer.hit(self.deck)
            self.dealer.show_hand()

        player_val = self.player.hand.get_value()
        dealer_val = self.dealer.hand.get_value()

        print("\n=== Result ===")
        if dealer_val > 21:
            print("Dealer busts! Player wins.")
        elif dealer_val > player_val:
            print("Dealer wins.")
        elif dealer_val < player_val:
            print("Player wins.")
        else:
            print("It's a tie!")

# ---- Run the game ----
if __name__ == "__main__":
    game = BlackjackGame()
    game.start()
```
